# A. Instrumentation DataFrame 
This problem asks me to create a DataFrame with 'Track == Intrumentation', 'Electronics > 70', and "Hometown == Luzon'

- The code will filter the students who follows the Instrumentation track, the students whose Hometown is Luzon, and the students who has scores that is greater than 70 in Electronics
- The output of the DataFrame will show three columns: Name, GEAS, and Electronics
- Conditions: 'Track == Instrumentation', 'Hometown == Luzon', and 'Electronics > 70'


```python
# Filtering the data for the instrumentation 'track', 'Electronics > 70', and Hometown = 'Luzon'
instru_df = df[(df['Track'] == 'Instrumentation') 
            &(df['Hometown'] == 'Luzon')
            &(df['Electronics'] 70)][['Name', 'GEAS', 'Electronics']] 

#Calling the filtered df 
instru_df
```

# B. Mindy DataFrame 
This problem asks me to create a DataFrame with 'Gender == Female', 'Hometown == Mindanao', and "Average >= 55'

- The code will filter the students who's Gender is Female, Hometown is Mindanao and Contains an Average greater than or equal to 55. 
- The output of the DataFrame will show four columns: Name, Track, Electronics and Average
- Conditions: 'Gender == Female', 'Hometown == Mindanao', and 'Average >= 55' 

```python
# Calculating for the average >=  55 
df['Average'] = df[['Math','Electronics', 'GEAS', 'Communication']].mean(axis = 1)

# Filtering the data for Mindy: 'track', 'Electronics', and 'Average >= 55'
mindy = df[(df['Hometown'] == 'Mindanao')
        &(df['Gender'] == 'Female')
        &(df['Average'] >= 55 )][['Name', 'Track', 'Electronics', 'Average']]

#Calling the filtered mindy
mindy
```

## 2. Creating Visualization that shows how the different features contributes to average grade.
   

```python
df
```

# Visualization using Bar Plot
- The output will be bar plots of Average grades in terms of Track and Gender

```python
# Importing the matplotlib.pyplot and seaborn libraries
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
# Calculating the average score for each student
df['Average'] = df[['Math','Electronics', 'GEAS', 'Communication']].mean(axis = 1)

# Setting up the figure size
plt.figure(figsize=(10,6))

# Plotting the average grades in terms of 'Track', 'Gender', and 'Hometown' 
sns.barplot(x ='Track', y ='Average', hue ='Gender', data=df)
plt.title('Average grade in terms of Track and Gender')

# Calling for the output 
plt.show()
```

# Visualization using Bar Plot 
- The output will show bar plots of Average grades in terms of Hometown and Track

```python
# Setting up the figure size
plt.figure(figsize=(10,6))

# Visualization for 'Hometown' comparison
sns.barplot(x = 'Hometown', y = 'Average', hue = 'Track', data=df) 
plt.title('Average grade in terms of Hometown and Track') 

# Calling for output
plt.show()
```

# Visualization using Box Plots
- The output will show box plots of Distribution of Average grades in terms of Track

```python
# Visualization using Box Plots

# Setting up the figure size
plt.figure(figsize=(10,6))

# Distribution of Average Grade in terms of Track
sns.boxplot(x = 'Track', y = 'Average', hue = 'Gender', data=df)
plt.title('Distribution of Average Grade in terms of Track')

# Calling for the output
plt.show()
```

# Visualization using Scatter Plot
- The output will show scatter plots of the relationship between the Scores in Electronics and their Average grade

```python
# Visualization using Scatter Plot

# Setting up the figure size
plt.figure(figsize=(10,6))

# Visualization using Scatter plot to see the relationship between Scores in Electronics and Average Grade
sns.scatterplot(x = 'Electronics', y = 'Average', hue = 'Gender', style = 'Track', data = df, s = 100)
plt.title('Scores in Electronics V.S Average Grade')

# Calling for the output
plt.show()
```

Does chosen track in college, gender, or hometown contributes to a higher average score?
- The small differences in average grades across the categories—track (Communication, Instrumentation, Microelectronics), gender (Female, Male), and hometown (Luzon, Mindanao, Visayas)—indicate that all groups achieved similar scores, with averages in the 60s. The graphs suggest that factors like track, gender, and hometown do not significantly affect average scores.
